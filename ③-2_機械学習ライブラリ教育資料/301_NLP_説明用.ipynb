```
def draw_scattter_plot_cluster(vecs,tags,cluster):

    colors = ["b","g","r","c","m","y","k"]

    pca = PCA(n_components=2)
    coords = pca.fit_transform(vecs)

    fig, ax = plt.subplots()

    print(ax)
    print(fig)

    x = [v[0] for v in coords]

    print(x)
    y = [v[1] for v in coords]
    print(y)
    c_count = max(cluster)

    for i in range(c_count + 1 ):
        xx = []
        yy = []
        for c,a,b,t in zip(cluster,x,y,tags):
            if i == c:
                xx.append(a)
                yy.append(b)
        ax.scatter(xx,yy,c= colors[i % 7]) 
    ax.legend(["0", "1", "2"])
    plt.show()


draw_scattter_plot_cluster(vecs, tags, cluster)
```
